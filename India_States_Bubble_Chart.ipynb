```python
Berozgaari = Pawan_pd.read_csv('/Users/brpawankumar/Desktop/India_States_Data/Umemployment__Rate.csv')
Berozgaari
```

```python
Ctr = 0
Mega_Merged_List =[] #This is the final list which we will use to create merged DataFrame
Unemployment = list(Berozgaari['Unemployment_Rate']) 
Selected_States = list(Bharat_Desh['st_nm'])
Bharat_Rajya = list(Berozgaari['States'])
#print("Literacy__Rate   ",Literacy__Rate)
#print("+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
#print("Selected_States   ",Selected_States)
#print("+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
#print("Bharat_Rajya   ",Bharat_Rajya)
#print("+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
for Rajya in Bharat_Rajya:
        Index_Wala_Array = Bharat_Desh.st_nm[Bharat_Desh.st_nm == Rajya].index.values
        Index_Wala_Number = Index_Wala_Array[0] #We retrive the index in Bharat_Desh based upon state name
        Mega_Merged_List.append([Index_Wala_Number,Rajya,Unemployment[Ctr],Bharat_Rajya[Ctr]])
        #Here we create a merged list that will be used to create a merged dataframe 
        Ctr = Ctr + 1
        
Merged_Dataframe = Pawan_pd.DataFrame(Mega_Merged_List,columns =['Index','States','Unemployment__Rate','Rajya'])
Merged_Dataframe.sort_values(by=['Index'], inplace=True)#Merged Dataframe is sorted as per 'Index' column
print(Merged_Dataframe)
```

```python
Merged_Dataframe.reset_index(inplace=True) #We reset the index so that generation of lists can be easy
print(Merged_Dataframe)
List_Main_Index = list(Merged_Dataframe['Index'])
List_Prep_Unemployment_Rate = list(Merged_Dataframe['Unemployment__Rate'])
```

```python
Final_Unemployment_Rate = []

Ctr = 0

while Ctr < len(Selected_States):
    Is_Element_Present = Ctr in List_Main_Index
    if Is_Element_Present == True:
        Temporary_Index = List_Main_Index.index(Ctr)
        Temporary_Unemployment_Rate = List_Prep_Unemployment_Rate[Temporary_Index ]
        Final_Unemployment_Rate.append(Temporary_Unemployment_Rate * 100)
    else:
        Final_Unemployment_Rate.append(0)
    Ctr = Ctr+ 1
print(Final_Unemployment_Rate)
```

```python
Bharat_Desh['Unemployment_Rate'] = Final_Unemployment_Rate
```

```python
Bharat_Desh
```

```python
fig, ax = Pawan_plt.subplots(figsize = (32, 24))
Pawan_plt.title("Plotting_Unempolyment_Rate_In_Indian_States")
Bharat_Desh.plot(ax = ax, column = 'Unemployment_Rate',cmap ='PuBuGn', edgecolor = 'white', linewidth = 0.5, legend = True)
ax.axis("off")
Pawan_plt.show()
```
